## Retriever Optimization Demo - Simple

```python
from ragbuilder.config.retriever import RetrievalOptionsConfig, BaseRetrieverConfig, RerankerConfig, RetrievalConfig
from ragbuilder.retriever.pipeline import RetrieverPipeline
from ragbuilder.retriever.optimization import run_optimization
from langchain_openai import AzureChatOpenAI, AzureOpenAIEmbeddings
```

```python
config = RetrievalOptionsConfig(
    retrievers=[
        BaseRetrieverConfig(
            type="similarity",
            retriever_k=[20],
            weight=0.7
        ),
        BaseRetrieverConfig(
            type="bm25",
            retriever_k=[20],
            weight=0.3
        )
    ],
    rerankers=[
        RerankerConfig(
            type="BAAI/bge-reranker-base",
            reranker_kwargs={}
        )
    ],
    top_k=[3, 5],
    optimization={
        "n_trials": 10, 
        "n_jobs": 1,
        "study_name": "lillog_agents_study",
        "optimization_direction": "maximize"
    },
    evaluation_config={
        "type": "ragas",
        "test_dataset": "rag_test_data_lilianweng_gpt-4o_1721032414.736622_SEMI.csv",
        "evaluator_kwargs": {
            "llm": AzureChatOpenAI(model="gpt-4o-mini", temperature=0.0),
            "embeddings": AzureOpenAIEmbeddings(model="text-embedding-3-small"),
        }
    },
    
    # Logging configuration
    log_config={
        "log_level": 20,  # INFO level
        "show_progress_bar": True,
        "verbose": True
    },
    
    database_logging=False,
    # database_path="lillog_agents_eval.db"
)
```

```python
import os
import getpass
os.environ["AZURE_OPENAI_API_KEY"] = getpass.getpass()
```

```python
results = run_optimization(config)
```

```python
best_config_dict=results["study"].user_attrs["trial_2_results"]["config"]
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain.retrievers import ParentDocumentRetriever
from langchain.storage import InMemoryStore
from ragbuilder.config.components import CHUNKER_MAP
```

```python
from ragbuilder.core.config_store import ConfigStore
BEST_CONFIG=ConfigStore.get_best_config()
```

```python
splitter_class = CHUNKER_MAP[BEST_CONFIG["chunking_strategy"]["type"]]()
```

```python
child_splitter = splitter_class(
    chunk_size=BEST_CONFIG["chunk_size"],
    chunk_overlap=BEST_CONFIG["chunk_overlap"],
    **BEST_CONFIG["chunking_strategy"]["chunker_kwargs"] or {}
)

parent_splitter = splitter_class(
    chunk_size=BEST_CONFIG["chunk_size"] * 3,
    chunk_overlap=BEST_CONFIG["chunk_overlap"] * 3,
    **BEST_CONFIG["chunking_strategy"]["chunker_kwargs"] or {}
)
```

```python
store = InMemoryStore()
retriever = ParentDocumentRetriever(
    vectorstore=best_indexer,
    docstore=store,
    child_splitter=child_splitter,
    parent_splitter=parent_splitter
)
```

```python
docs = STORE.get_best_config_docs()
```

```python
retriever.add_documents(docs[0])
```

```python
results = retriever.invoke(query)
```
